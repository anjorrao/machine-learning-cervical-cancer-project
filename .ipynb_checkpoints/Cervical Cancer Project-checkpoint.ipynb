# Splitting dataset into dependent and independent variables.

```python
X = df.iloc[:,:-4].values
y = df.iloc[:,-4].values
```

# Splitting training and test data

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.25, random_state = 0)
```

# Handling class imbalance problem

```python
sns.countplot(x=y_train)
```

From the above graph, the target column 'Hinselmann' is indicating class imbalance problem. Thus we are handling this problem using SMOTE technique.
SMOTE (Synthetic Minority Oversampling Technique) is oversampling technique where synthetic samples of minority class is created.

```python
sm = SVMSMOTE(random_state=42)
X_train_over, y_train_over = sm.fit_resample(X_train, y_train)
```

```python
print('Before SMOTE')
print(Counter(y_train))
print('================================ \nAfter SMOTE')
print(Counter(y_train_over))
```

```python
sns.countplot(x=y_train_over)
```

# Feature scaling

```python
sc = StandardScaler()
X_train_over = sc.fit_transform(X_train_over)
X_test = sc.transform(X_test)
```

# Training the Logistic regression model

```python
logreg = LogisticRegression(random_state = 0)
logreg.fit(X_train_over,y_train_over)
```

#  Predicting test set results (Logistic regression)

```python
y_pred = logreg.predict(X_test)
```

# Making confusion matrix (Logistic regression)

```python
logcm = confusion_matrix(y_test,y_pred)

ax= plt.subplot()
sns.heatmap(logcm, annot=True, fmt='g', ax=ax);  #annot=True to annotate cells, ftm='g' to disable scientific notation

# labels, title and ticks
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels'); 
ax.set_title('Confusion Matrix'); 
ax.xaxis.set_ticklabels(['Cancer', 'Non-cancer']); ax.yaxis.set_ticklabels(['Cancer', 'Non-cancer']);


print("Accuracy score ", accuracy_score(y_test, y_pred))
print("Precision score ", precision_score(y_test, y_pred))
print("Recall score", recall_score(y_test, y_pred))
print("F1 score", f1_score(y_test, y_pred))
```

# Plotting ROC Curve (Logistic Regression Model)

```python
#define metrics
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred)

#create ROC curve
plt.plot(fpr,tpr)
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
